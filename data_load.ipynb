```python
from kagglehub import KaggleDatasetAdapter, load_dataset
import pandas as pd
import numpy as np
```

# ***Audio Feature Data Preprocessing***

```python
def rename_cols(raw_df):
    feature_names = raw_df.iloc[:2,:]
    scaled_feature_names = ['track_id']

    # extract custom column names
    for col in feature_names.columns[1:]:
        feature_type = feature_names[col][0]
        feature_stat = feature_names[col][1]
        col_name = col.split('.')[0]
        new_col_name = f"{col_name}_{feature_type}_{feature_stat}"
        scaled_feature_names.append(new_col_name)   

    # create new dataframe with renamed columns
    scaled_df = raw_df.iloc[3:, :]
    scaled_df.columns = scaled_feature_names
    scaled_df.reset_index(drop=True, inplace=True)

    # convert all columns from object to numeric
    scaled_df = scaled_df.apply(pd.to_numeric)

    return scaled_df
```

```python
file_path = "fma_metadata/features.csv"
audio_features = load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "imsparsh/fma-free-music-archive-small-medium",
  file_path,
)
```

```python
audio_features.head()
```

```python
structured_audio_features = rename_cols(audio_features)
structured_audio_features
```

```python
structured_audio_features.info()
```

# ***Title Extraction For Each tracks***

```python
file_path = "fma_metadata/tracks.csv"
song_tracks = load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "imsparsh/fma-free-music-archive-small-medium",
  file_path,
)
```

```python
def extract_titles (tracks_df):
    # extract song titles from tracks dataframe
    titles = pd.DataFrame(tracks_df.iloc[:, -1])
    titles = titles.iloc[2:, :]
    titles.columns = ['title']
    titles.reset_index(drop=True, inplace=True)

    # convert title df from object to string type
    titles['title'] = titles['title'].astype(str)
    return titles
```

```python
song_tracks.head()
```

```python
tracks_with_titles = extract_titles(song_tracks)
tracks_with_titles
```

```python
tracks_with_titles.to_csv('./dataset/music_tracks_with_titles.csv', index=False)
```

```python
data = pd.concat([tracks_with_titles,structured_audio_features], axis=1)
# save the processed data to a csv file
data.to_csv('./dataset/music_data_with_audio_feat.csv', index=False)
data
```

```python
from lyricsgenius import Genius
import time
import logging
```

```python
def get_lyrics(song_df):
    logging.getLogger("lyricsgenius").setLevel(logging.ERROR)
    lyrics_list = []
    genius = Genius(
        CLIENT_TOKEN,
        timeout=10,
        retries=3,
        remove_section_headers=True,
        skip_non_songs=True
    )

    a = 0
    got = 0
    for title in song_df['title']:
        try:
            song = genius.search_song(title)

            if song and song.lyrics:
                lyrics_list.append(song.lyrics)
                got += 1
            else:
                lyrics_list.append(np.nan)

            time.sleep(1)   # prevents rate limit
            print(f"{a+1}/{len(song_df)} & {got} lyrics found", end='\r')
            a += 1

            
        except Exception as e:
            lyrics_list.append(np.nan)

    song_df['lyrics'] = lyrics_list
    
    # convert lyrics column to string type
    song_df['lyrics'] = song_df['lyrics'].astype('string')
    return song_df
```
